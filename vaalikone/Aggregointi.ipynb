```python
from collections import defaultdict
import ylekone
```

```python
vk = ylekone.Vaalikone()
municipality = vk["Helsinki"]
stats = ylekone.make_stats(municipality)
```

```python
def max_disagreement(stats, party_a, party_b):
    entries = []
    for qid, statsq in stats.items():
        a = statsq[party_a]
        b = statsq[party_b]
        delta = a.mean - b.mean
        entries.append((abs(delta), delta, qid, a, b))
    entries = list(reversed(sorted(entries)))
    return entries
```

```python
e1 = max_disagreement(stats, "Vihr.", "Vas.")
e2 = max_disagreement(stats, "Vihr.", "SDP")
e = list(reversed(sorted(e1 + e2)))

for _, delta, qid, a, b in e:
    print(f"Δ {delta:+.2f}: {municipality.question_id_to_text(qid)} ({qid})")
    print(f"  {a.prettystr}")
    print(f"  {b.prettystr}")
    print()
```
